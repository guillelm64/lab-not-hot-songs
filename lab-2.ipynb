```python
### Now we will generate a dataframe with more or less 10,000 random songs. 
### This dataframe was obtained from the Kaggle website https://www.kaggle.com/datasets/sumitmohod22/songs-data-set?resource=download

random_songs = pd.read_csv('song_data.csv')
display(random_songs)
print(random_songs.shape)
```

```python
### This function generated a dataframe with only two columns, song and artist, drops duplicates
### and searches for matches between the top_100 songs and the random_songs

def generate_random_songs(x, y):
    y = y[['title', 'artist']]
    y.columns = ["song", "artist"]
    y = y.drop_duplicates( subset = ['song', 'artist']).reset_index(drop=True) 
    y = y.sample(n=3000, replace=False).reset_index(drop=True)
    for i in range(len(x)):
        for j in range(len(y)):
            if list(x.iloc[i]) == list(y.iloc[j]):
                y = y.drop(y.iloc[j], axis = 0)
            else:
                pass                 
    return y
```

```python
random_songs = generate_random_songs(top_100, random_songs)
display(random_songs)
random_songs.shape
```
